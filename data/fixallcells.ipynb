```python
import pandas as pd
```

```python
all_cells_file = "all_cells.txt"
all_cells_fixed = "all_cells.csv"

with open(all_cells_file, "r") as f:
    counter = 0
    curr_row_counter = 0
    for line in f.readlines():
        if "From Sulston, JE and White, JG (1988)" in line:
            break
        if counter == 4:
            columns_dict = {title.strip():[] for title in line.split('\t')}
        if counter > 4:
            if curr_row_counter == 0:
                columns_dict['Cell'].append(line.strip())
            elif curr_row_counter == 2:
                columns_dict['Lineage Name'].append(line.strip())
            elif curr_row_counter == 4:
                columns_dict['Description'].append(line.strip())
            elif curr_row_counter > 4:
                curr_row_counter = -1
            curr_row_counter += 1
        counter += 1

print(columns_dict)
print({colname:len(data) for (colname, data) in columns_dict.items()})

df = pd.DataFrame.from_dict(columns_dict)
df.to_csv(all_cells_fixed, index=False)
```
