```python
from owmeta_core.command import OWM
conn = OWM().connect()

from owmeta_core.context import Context
ctx = conn(Context)(ident='http://openworm.org/data')

from owmeta.worm import Worm
net = ctx.stored(Worm).query().neuron_network()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import pandas as pd
import seaborn as sns
```

```python
import os
import re
```

```python
def sort_key(s):
    return [int(i) if i.isdigit() else i for i in re.split(r'(\d+)', s)]
```

```python
neuron_to_receptors_dict = {'neuron':[], 'receptor':[]}
neurons = list(net.neurons())
neurons.sort()
receptors = dict()
for neuron in neurons:
    new_receptors = list(neuron.receptors())
    new_receptors.sort(key=sort_key)
    for receptor in new_receptors:
        neuron_to_receptors_dict['neuron'].append(neuron.name())
        neuron_to_receptors_dict['receptor'].append(receptor)
neuron_to_receptors_df = pd.DataFrame.from_dict(neuron_to_receptors_dict)
neuron_to_receptors_df.to_csv(os.path.join("data", "neuron_to_receptors.csv"), index=False)
neuron_to_receptors_df
```

```python
receptors = set()
for neuron in neurons:
    receptors = receptors | set(neuron.receptors())
receptors = list(receptors)
receptors.sort(key=sort_key)

with open(os.path.join('data', 'receptors.txt'), 'w') as f:
    for receptor in receptors:
        f.write(f"{receptor}\n")
```

```python
# Iterate over each neuron to retrieve its neurotransmitter(s)
neuron_to_nts_dict = {'neuron':[], 'neurotransmitter':[]}
for neuron in neurons:
    new_neurotransmitters = list(neuron.neurotransmitter())
    new_neurotransmitters.sort()
    for neurotransmitter in new_neurotransmitters:
        neuron_to_nts_dict['neuron'].append(neuron.name())
        neuron_to_nts_dict['neurotransmitter'].append(neurotransmitter)
neuron_to_nts_df = pd.DataFrame.from_dict(neuron_to_nts_dict)
neuron_to_nts_df.to_csv(os.path.join("data", "neuron_to_neurotransmitters.csv"), index=False)
neuron_to_nts_df
```

```python
nts = set()
for neuron in neurons:
    nts = nts | set(neuron.neurotransmitter())
nts = list(nts)
nts.sort()

with open(os.path.join('data', 'neurotransmitters.txt'), 'w') as f:
    for nt in nts:
        f.write(f"{nt}\n")
```

```python
# Iterate over each neuron to retrieve its neuropeptide(s)
neuron_to_peptides_dict = {'neuron':[], 'neuropeptide':[]}
for neuron in neurons:
    new_neuropeptides = list(neuron.neuropeptide())
    new_neuropeptides.sort(key=sort_key)
    for neuropeptide in new_neuropeptides:
        neuron_to_peptides_dict['neuron'].append(neuron.name())
        neuron_to_peptides_dict['neuropeptide'].append(neuropeptide)
neuron_to_peptides_df = pd.DataFrame.from_dict(neuron_to_peptides_dict)
neuron_to_peptides_df.to_csv(os.path.join("data", "neuron_to_neuropeptides.csv"), index=False)        
neuron_to_peptides_df
```

```python
nps = set()
for neuron in neurons:
    nps = nps | set(neuron.neuropeptide())
nps = list(nps)
nps.sort(key=sort_key)

with open(os.path.join('data', 'neuropeptides.txt'), 'w') as f:
    for np in nps:
        f.write(f"{np}\n")
```

```python
# Iterate over each neuron to retrieve its channel(s)
neuron_to_channels_dict = {'neuron':[], 'channel':[]}
for neuron in neurons:
    new_channels = list([channel.name() for channel in neuron.channel()])
    new_channels.sort(key=sort_key)
    for channel in new_channels:
        neuron_to_channels_dict['neuron'].append(neuron.name())
        neuron_to_channels_dict['channel'].append(channel)
neuron_to_channels_df = pd.DataFrame.from_dict(neuron_to_channels_dict)
neuron_to_channels_df.to_csv(os.path.join("data", "neuron_to_channels.csv"), index=False)
neuron_to_channels_df  
```
